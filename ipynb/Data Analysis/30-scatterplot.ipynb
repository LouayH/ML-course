```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("datasets/world-happiness.csv")

data.head()
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(data = data, x = "Economy (GDP per Capita)", y = "Happiness Score", hue = "Region")
```

```python
grid = sns.FacetGrid(data, col = "Region", hue = "Region", col_wrap = 5)
grid.add_legend()
grid.map(sns.scatterplot, "Economy (GDP per Capita)", "Health (Life Expectancy)")
```
