```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv("datasets/car-malfunctions.csv")

dataframe.info()
```

```python
yValues = dataframe["malfunctions"]
plt.plot(yValues, marker = 'o')
```

```python
malfunctions = dataframe["malfunctions"]
age = dataframe["Age"]

plt.plot(malfunctions, marker = 'o')
plt.plot(age, marker = 'o')
```

```python
xValues = ["Toyota", "BMW", "Mercedes", "Kia", "Hunday"]
yValues = [15, 30, 40, 55, 60]

plt.xlabel("Name")
plt.ylabel("Malfunctions")
plt.bar(xValues, yValues, width = 0.5)
```

```python
plt.barh(xValues, yValues, height = 0.5)
```
