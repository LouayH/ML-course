```python
import matplotlib.pyplot as plt
```

```python
data = [1, 2, 3]
```

```python
# 1, 2, 3 are y axis values
# if not given, x axis values will be data length - 1
plt.plot(data)
```

```python
xValues = [2, 5, 6]
plt.plot(xValues, data)
```

```python
plt.title("matplotlib", fontsize = 12)
plt.xlabel("X Axis", fontsize = 8)
plt.ylabel("Y Axis", fontsize = 8)
plt.plot(xValues, data, linewidth = 3, marker = "o", ls = ":", color="r")
```
