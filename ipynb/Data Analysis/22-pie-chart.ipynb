```python
import matplotlib.pyplot as plt
```

```python
data = [35, 25, 25, 15]

plt.pie(data)
```

```python
labels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(data, labels = labels)
```

```python
explodes = [0.05, 0.05, 0, 0]

plt.pie(data, labels = labels, explode = explodes, shadow=True)
```

```python
colors = ["Gold", "Yellow", "Red", "Brown"]

plt.pie(data, labels = labels, colors = colors)
```

```python
plt.pie(data, labels = labels)
plt.legend()
```
