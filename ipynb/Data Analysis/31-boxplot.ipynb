```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset("tips")

data.head()
```

```python
data.boxplot(by = "day", column = ["total_bill"], grid = False)
```

```python
data.boxplot(by = "size", column = ["tip"], grid = False)
```

```python
sns.boxplot(x = "day", y = "tip", data = data, hue = "day", palette="hls")
```
