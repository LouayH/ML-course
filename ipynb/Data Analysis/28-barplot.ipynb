```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset("titanic")

data.head()
```

```python
sns.barplot(x = "sex", y = "survived", data = data)
```

```python
sns.barplot(x = "survived", y = "sex", data = data)
```

```python
sns.barplot(x = "survived", y = "sex", data = data, color = "brown")
```

```python
sns.barplot(x = "survived", y = "sex", data = data, hue = "sex", palette = "hls")
```

```python
sns.barplot(x = "class", y = "survived", data = data, hue = "embark_town")
```

```python
sns.barplot(x = "class", y = "survived", data = data, hue = "embark_town", errorbar = None)
```

```python
sns.barplot(x = "class", y = "survived", data = data, hue = "embark_town", order = ["Third", "Second", "First"])
```
