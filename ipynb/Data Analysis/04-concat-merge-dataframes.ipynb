```python
import pandas as pd
```

```python
dataframe_1 = pd.DataFrame({
  "#1": [1, 2, 3, 4],
  "#2": ["A", "B", "C", "D"],
  "#3": [True, False, True, False]
})

dataframe_1
```

```python
dataframe_2 = pd.DataFrame({
  "#1": [3, 4, 5, 6],
  "#2": ["E", "D", "G", "H"],
  "#3": [False, True, False, True]
})

dataframe_2
```

```python
pd.concat([dataframe_1, dataframe_2])
```

```python
pd.concat([dataframe_1, dataframe_2], keys=['df1', 'df2'])
```

```python
pd.concat([dataframe_1, dataframe_2], ignore_index=True)
```

```python
pd.concat([dataframe_1, dataframe_2], axis=1)
```
