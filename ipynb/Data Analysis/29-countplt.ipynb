```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset("tips")

data.head()
```

```python
sns.countplot(x = "sex", data = data)
```

```python
sns.countplot(data = data, x = "day", hue = "day")
```

```python
sns.countplot(data = data, x = "day", hue = "day", order = data["day"].value_counts().index)
```

```python
sns.countplot(data = data, x = "day", hue = "day", order = data["day"].value_counts(ascending=True).index)
```

```python
sns.countplot(data = data, x = "sex", hue = "smoker")
```

```python
subplot = sns.countplot(data = data, x = "sex", hue = "smoker")

for label in subplot.containers:
  subplot.bar_label(label)
```

```python
sns.countplot(data = data, y = "sex", hue = "smoker")
```

```python
sns.countplot(data = data, x = "sex", hue = "sex", palette="Set2")
```
