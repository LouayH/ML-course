```python
import pandas as pd
```

```python
dataframe = pd.DataFrame({
  "city": ["A", "B", "C"],
  "day1": [22, 25, 28],
  "day2": [10, 14, 13],
  "day3": [25, 22, 26],
  "day4": [18, 15, 14],
  "day5": [17, 14, 18],
})

dataframe
```

```python
dataframe.melt(id_vars=["city"])
```

```python
dataframe.melt(id_vars=["city"], var_name="day", value_name="temperature")
```

```python
dataframe.melt(id_vars=["city"], var_name="day", value_name="temperature").sort_values(by="city")
```

```python
dataframe.melt(id_vars=["city"], var_name="day", value_name="temperature").sort_values(by="city").reset_index(drop=True)
```
