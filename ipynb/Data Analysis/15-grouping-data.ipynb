```python
dataframe = dataframe[dataframe["Quantity"] >= 0]

dataframe = dataframe.assign(Total=dataframe.Quantity * dataframe.UnitPrice)

dataframe
```

```python
invoices_by_country_customer = dataframe.groupby(["Country", "CustomerID"])
invoices_by_country_customer.groups
```

```python
invoices_by_country_customer.count()
```

```python
total_by_country_customer = invoices_by_country_customer.agg(
  TotalPrice=("Total", "sum"),
)

total_by_country_customer
```

```python
total_by_country_customer.sort_values(by=["TotalPrice"], ascending=False)
```
