```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv("datasets/car-malfunctions.csv")

dataframe.info()
```

```python
age = dataframe["Age"]
malfunctions = dataframe["malfunctions"]
daysOff = dataframe["DaysOff"]

# subplot is in a grid of 1 row, 2 columns (1st plot)
plt.subplot(1, 2, 1)
plt.plot(age, malfunctions, marker = "o")

# (2nd plot)
plt.subplot(1, 2, 2)
plt.plot(malfunctions, age, marker = "o")
```

```python
# subplot is in a grid of 2 rows, 1 column (1st plot)
plt.subplot(2, 1, 1)
plt.plot(age, malfunctions, marker = "o")

# (2nd plot)
plt.subplot(2, 1, 2)
plt.plot(malfunctions, age, marker = "o")
```

```python
plt.subplot(1, 3, 1)
plt.plot(age, malfunctions, marker = "o")

plt.subplot(1, 3, 2)
plt.plot(malfunctions, age, marker = "o")

plt.subplot(1, 3, 3)
plt.plot(age, daysOff, marker = "o")
```

```python
plt.subplot(3, 3, 1)
plt.plot(age, age)

plt.subplot(3, 3, 2)
plt.plot(age, malfunctions)

plt.subplot(3, 3, 3)
plt.plot(age, daysOff)

plt.subplot(3, 3, 4)
plt.plot(malfunctions, age)

plt.subplot(3, 3, 5)
plt.plot(malfunctions, malfunctions)

plt.subplot(3, 3, 6)
plt.plot(malfunctions, daysOff)

plt.subplot(3, 3, 7)
plt.plot(daysOff, age)

plt.subplot(3, 3, 8)
plt.plot(daysOff, malfunctions)

plt.subplot(3, 3, 9)
plt.plot(daysOff, daysOff)
```
