```python
from keras.datasets import imdb

(x_train, y_train), (x_test, y_test) = imdb.load_data()

len(x_train), len(x_test)
```

The most frequent words in the dataset

```python
word_index = imdb.get_word_index()

word_index = dict(sorted(word_index.items(), key=lambda item: item[1]))

word_index
```

First review in the training dataset is labeled **positive** and has **218** words indexed by overall frequency in the dataset.

```python
print(len(x_train[0]), x_train[0], y_train[0])
```

Second review in the training dataset is labeled **negative** and has **189** words indexed by overall frequency in the dataset.

```python
print(len(x_train[1]), x_train[1], y_train[1])
```

```python
# swap key-value pairs in word_index dictionary
word_index = { v: k for k, v in word_index.items()}

# get the 11th most frequent word in the dataset
word_index.get(11)
```

```python
positive_indices = []
negative_indices = []

for i in range(len(y_train)):
  positive_indices.append(i) if y_train[i] == 1 else negative_indices.append(i)
```

```python
review_word_indexes = x_train[positive_indices[0]]

# the first 3 indexes are reserved in the imdb dataset
# for example: "the" is the most frequent word, but actually its index is 4
review_words = [word_index.get(idx - 3, "") for idx in review_word_indexes]

positive_review = " ".join(review_words).strip()

positive_review
```

```python
review_word_indexes = x_train[negative_indices[0]]

# the first 3 indexes are reserved in the imdb dataset
# for example: "the" is the most frequent word, but actually its index is 4
review_words = [word_index.get(idx - 3, "") for idx in review_word_indexes]

negative_review = " ".join(review_words).strip()

negative_review
```
