```python
import pandas as pd

data = pd.read_csv("datasets/vip.csv")

data.head()
```

```python
x = data[["Age", "Education"]]
y = data["VIP"]
```

```python
VIP_data = data[data["VIP"] == 1]

VIP_data.head()
```

```python
NonVIP_data = data[data["VIP"] == 0]

NonVIP_data.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(VIP_data["Education"], VIP_data["Age"], color = "red", marker = "x")
plt.scatter(NonVIP_data["Education"], NonVIP_data["Age"], color = "blue", marker = "x")

plt.legend(["VIP", "Non-VIP"])
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x, y)

intercept = model.intercept_
coefficient = model.coef_

print ("y = ", intercept, "+", coefficient[0], "* Age", "+", coefficient[1], "* Eduction")
```

```python
import math

def sigmoid(y):
  return 1 / (1 + pow(math.e, -y))

age = 69
education = 12

decision_boundry = 0.5

y_predict = intercept + coefficient[0] * age + coefficient[1] * education

y_predict = sigmoid(y_predict)

"VIP" if y_predict > decision_boundry else "Non-VIP"
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(x.values, y)
```

```python
sample_data = [[69, 12]]

y_predict = model.predict(sample_data)

"VIP" if y_predict == 1 else "Non-VIP"
```

```python
model.predict_proba(sample_data)

# probabilities of being Non-VIP vs VIP
```
